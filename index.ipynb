The best model seems to be the 1st order Lasso regression with alpha = 3.68, this was chosen for its high adjusted r squared (test) as well as its high 5-fold cv  
We the run the best performing model in order to get insight into the residuals

```python
X_train, X_test, y_train, y_test= train_test(diamonds, ['carat','x','y','z','depth','table'], 'price')    

    
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
model_lass = LassoCV(max_iter = 100000, cv = 5)

model_lass.fit(X_train, y_train)

pred_lass = model_lass.predict(X_test)
```

```python
model_lass.coef_
```

```python
lst =(pred_lass - y_test)
```

```python
len(lst)
```

```python
sns.set(style = 'darkgrid', font_scale=2)
plt.figure(figsize = (18,10))
ax = sns.regplot(x= y_test ,y = (lst))
ax.set(xlabel = 'Price', ylabel = 'Residual', title = 'Residual Plot for Lasso - Diamond Price');
```

The 4th order Ridge regression with an alpha of 9 (could be higher we stopped testing) seems to perform the best with the highest combination of adjusted r squared (test) and 5 fold cv

```python
X_train, X_test, y_train, y_test= train_test(life_exp, ['alcohol', 'hiv', 'gdp', 'school', 'death_ratio'], 'life_expectancy')    

    
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
model_ridge = RidgeCV(alphas=[9], cv = 5)

model_ridge.fit(X_train, y_train)

pred_ridge = model_ridge.predict(X_test)
```

```python
lst_rdg = (pred_ridge - y_test)
lst_rdg[:10]
```

```python
len(pred_ridge), len(y_test)
```

```python
plt.figure(figsize = (18,10))
ax = sns.regplot(x= y_test ,y = (lst_rdg[::1]))
ax.set(xlabel = 'Life Expectancy', ylabel = 'Residual', title = 'Residual Plot for Ridge - Life Expectancy');
```
