```python
X_train, X_test, y_train, y_test= train_test(diamonds, ['carat','x','y','z','depth','table'], 'price')    

    
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
model_lass = LassoCV(max_iter = 100000, cv = 5)

model_lass.fit(X_train, y_train)

pred_lass = model_lass.predict(X_test)
```

```python
lst =(pred_lass - y_test)
```

```python
len(lst)
```

```python
sns.set(style = 'darkgrid', font_scale=2)
plt.figure(figsize = (18,10))
ax = sns.regplot(x= y_test ,y = (lst))
ax.set(xlabel = 'Price', ylabel = 'Residual', title = 'Residual Plot for Lasso');
```

```python
X_train, X_test, y_train, y_test= train_test(life_exp, ['alcohol', 'hiv', 'gdp', 'school', 'death_ratio'], 'life_expectancy')    

    
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
model_ridge = RidgeCV(alphas=[9], cv = 5)

model_ridge.fit(X_train, y_train)

pred_ridge = model_ridge.predict(X_test)
```

```python
lst_rdg = (pred_ridge - y_test)
lst_rdg[::1]
```
